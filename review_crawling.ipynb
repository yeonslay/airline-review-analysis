```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import random
import pandas as pd
```

```python
all_reviews = []

while True:
    time.sleep(2)
    # 리뷰블록단위로 모든 요소 찾기
    review_blocks = driver.find_elements(By.CSS_SELECTOR, ".yotpo-review.yotpo-regular-box")

    # 블록별로 본문과 날짜 추출
    for block in review_blocks:
        
        # 불필요한 항목이 포함되는 것을 방지하기 위한 조건문
        review_id = block.get_attribute("data-review-id")
        if review_id == "0":
            continue
            
        # 본문 추출
        content = block.find_element(By.CLASS_NAME, "content-review").text.strip()
        
        # 날짜 추출
        date = block.find_element(By.CLASS_NAME, "yotpo-review-date").text.strip()

        # 수집한 데이터 추가
        all_reviews.append({
            "date": date,
            "review": content
        })

    # 다음 페이지 이동
    try:
        # 버튼 찾기
        next_btn = driver.find_element(By.CSS_SELECTOR, 'a[aria-label="Next Page"]')

        # 마지막 페이지에서의 오류를 없애기 위한 조건문
        is_disabled = next_btn.get_attribute("aria-disabled")
        if is_disabled == "true":
            print("마지막 페이지 도달! 크롤링 종료!!")
            break

        # 페이지 이동 버튼 클릭
        next_btn.click()
        time.sleep(3)

    except Exception as e:
        print("다음 버튼을 찾을 수 없거나 클릭 실패:", e)
        break
```
